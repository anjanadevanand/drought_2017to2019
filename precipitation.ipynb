```python
import xarray as xr
import numpy as np
import pandas as pd
import os

import sys
new_path = '/home/566/ad9701/lulcc_ce_analysis/'
if new_path not in sys.path:
    sys.path.append(new_path)


import spi
nspi = 3

%run -i drought_2017to2020/precip_alldata_info.py

out_dir = '/g/data/w97/ad9701/drought_2017to2020/precip_data/'
data_names = list(alldata_dict.keys())
latlon_rename = {'latitude': 'lat', 'longitude': 'lon'}

for d in [data_names[2]]:
    star_loc = alldata_dict[d]['file_name'].find('*')
    if star_loc == -1:
        ds = xr.open_dataset(alldata_dict[d]['full_path'] + alldata_dict[d]['file_name'])
    else:
        ds = xr.open_mfdataset(alldata_dict[d]['full_path'] + alldata_dict[d]['file_name'])
        
    if alldata_dict[d]['rename_latlon']:
        da_AU = ds[alldata_dict[d]['varname']].rename(latlon_rename).sel(lat = alldata_dict[d]['lat_slice'], lon = alldata_dict[d]['lon_slice'])
    else:
        da_AU = ds[alldata_dict[d]['varname']].sel(lat = alldata_dict[d]['lat_slice'], lon = alldata_dict[d]['lon_slice'])
    
    out_file = d + '_monthly_' + str(da_AU['time.year'].min().values) + '_' + str(da_AU['time.year'].max().values) + '.nc'
    da_AU.to_netcdf(out_dir + out_file)
    
    rfile = out_dir + out_file
    wfile = out_dir + 'spi_' + str(nspi) + 'm_' + out_file
    vname = alldata_dict[d]['varname']
    x=spi.spi_netcdf_grid(rfile, wfile, vname, nspi)
    print(d + ' done!')
    
```
