### Save daily data

```python
import glob

years = list(range(2008, 2021))
lat_slice = slice(-10, -44)
lon_slice = slice(112, 154)

out_dir = '/g/data/w97/ad9701/drought_2017to2020/sm_data/daily_data/'
for yr in years:
    daily_files = sorted(glob.glob('/g/data/w97/ad9701/ESACCI/3_unzip/' + str(yr) + '/ESACCI-*.nc'))
    ds = xr.open_mfdataset(daily_files)
    da_sm = ds['sm'].sel(lat = lat_slice, lon = lon_slice)
    da_sm_uncertainty = ds['sm_uncertainty'].sel(lat = lat_slice, lon = lon_slice)
    sm_file = out_dir + 'esacci_sm_' + str(yr) + '.nc'
    da_sm.to_netcdf(sm_file)
    sm_uncert_file = out_dir + 'esacci_sm_uncertainty_' + str(yr) + '.nc'
    da_sm_uncertainty.to_netcdf(sm_uncert_file)
```
